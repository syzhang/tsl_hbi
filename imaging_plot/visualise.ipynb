### visualise imaging results

```python
import nilearn.plotting
```

#### all pain

```python
img = '../../TSL_output/2ndLevel_masked_FWE005/2ndLevel/con_0001/spmT_0001_thr.nii'
ttl = 'all_pain (FWE p<0.05)'
```

```python
g = nilearn.plotting.plot_stat_map(
    img, title=ttl,
    threshold=4, vmax=10, display_mode='z', cut_coords=(-15, -5, 0, 5, 15),
    colorbar=True);
```

```python
g = nilearn.plotting.plot_stat_map(
    img, title=ttl,
    threshold=4, vmax=10, display_mode='x', cut_coords=(-45, -30, -15, 0),
    colorbar=True);
```

```python
g.savefig('./figs/all_pain_x.pdf')
```

#### high>low pain

```python
img = '../../TSL_output/2ndLevel_masked_FDR0001/2ndLevel/con_0004/spmT_0001_thr.nii'
ttl = 'high>low pain (p<0.001)'
```
